```python
df1 = df.dropna(how='all', axis=1)
print(df1)
```

```python
df1 = df.dropna(subset=['Rating'])
print(df1)
```

```python
df2 = df1.drop(df1.index[2])
print(df2)
```

```python
#since we know only ratong and reviews are float and missing values of ratings are already removed. 
#Therefore to fill missing value of indiviual colums we will use mode of that columns

df2.isnull().sum()
```

```python
df2.fillna(df2.mode(), inplace=True)
```

```python
df2 = df2.apply(lambda x: x.fillna(x.value_counts().index[0]))
```
